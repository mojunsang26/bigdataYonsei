```python
try :
    driver.get("lala.com") # 에러가 발생할 가능성이 있는 코드
except :  # 에러 종류
    print('url이 잘못되었거나 서버가 불안정합니다')  #에러가 발생 했을 경우 처리할 코드
```

```python
import time
treeHit = 0
while treeHit < 3:
    treeHit = treeHit +1
    try :
        driver.get("lala.com") # 에러가 발생할 가능성이 있는 코드
    except :  # 에러 종류
        print('url이 잘못되었거나 서버가 불안정합니다\n1초 대기 후 재접합니다')  #에러가 발생 했을 경우 처리할 코드
        time.sleep(1)
    if treeHit == 3:
        print("3회 시도했습니다. 시도 중지")
```
