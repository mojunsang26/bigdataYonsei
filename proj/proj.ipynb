```python
import pandas as pd
import numpy as np

rawData = pd.read_csv('data.csv', encoding='CP949')
rawData.head()
```

```python
myData = rawData.loc[:,'기준년도':'흡연상태']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = myData.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
sns.plt.show()
```

```python
corr
```

```python
myCor = []
myidx = []
for i in range(26):
    for j in range(0,i):
        myCor.append(abs(float(corr.iloc[i,j])))
        myidx.append(corr.index[i] + ", " + corr.index[j])
wow = pd.DataFrame()
wow['index'] = myidx
wow['abs corr'] = myCor
wow.sort_values(by = 'abs corr', ascending = False)
```
