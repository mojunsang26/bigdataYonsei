```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
# import csv
```

```python
data = pd.read_csv('건강검진정보-1.csv', engine = 'python')
data.head()
```

```python
data = data.set_index('가입자일련번호')
```

```python
data_male = data[data['성별코드']==1]
data_male
```

```python
data_female = data[data['성별코드']==2]
data_female.head()
```

```python
data_male['신장'].head()
```

```python
np.std(data_male['신장'])
```

```python
표준신장 = (data_male['신장']-np.mean(data_male['신장']))/np.std(data_male['신장'])
```

```python
표준체중 = (data_male['체중']-np.mean(data_male['체중']))/np.std(data_male['체중'])
```

```python
표준허리둘레 = (data_male['허리둘레']-np.mean(data_male['허리둘레']))/np.std(data_male['허리둘레'])
```

```python
data_male['신장'] = 표준신장 + 10
data_male['체중'] = 표준체중 + 10
data_male['허리둘레'] = 표준허리둘레 + 10
```

```python
data_male
```

```python
data_male = data_male.dropna()
data_male
```
