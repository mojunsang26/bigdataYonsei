# 영화 평점 분석 실습

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

## 1. 영화 평점 데이터 적재 및 전처리

```python
#사용자 데이터 불러오기
users = pd.read_csv('data/movielens/users.dat', sep = '::', engine = 'python',
                    names = ['사용자아이디','성별','연령','직업','지역'])
users.head()
```

```python
#평점 데이터 불러오기
ratings = pd.read_csv('data/movielens/ratings.dat', sep = '::', engine = 'python',
                    names = ['사용자아이디','영화아이디','평점','타임스탬프'])
ratings.head()
```

```python
len(ratings)
```

```python
#영화 데이터 불러오기(movies.dat)
#컬럼명 : ['영화아이디','영화제목','장르']
movies = pd.read_csv('data/movielens/movies.dat',sep = '::', engine = 'python',
                    names = ['영화아이디','영화제목','장르'])
movies.head()
```

```python
temp = pd.merge(users, ratings)
data = pd.merge(temp, movies)
```

```python
data.head()
```

## 2. 보고 싶은 영화  찾기 (평점 높은 영화 찾기)

```python
data.pivot_table(index = '영화제목', aggfunc = 'mean', values = '평점').sort_values(by = '평점', ascending = False).head(15)
```

```python
영화평점 = data.pivot_table(index = '영화제목', aggfunc = ('mean','count'), values = '평점').sort_values(by = 'mean', ascending = False)
영화평점
```

```python
#1. 평점 개수가 가장 많은 영화 상위 10개 찾기
영화평점.sort_values(by = 'count', ascending = False)
영화평점.head(10)
```

```python
#2. 평점이 4.5점 이상인 영화들 중에 평점의 개수가 많은 순으로 정렬해서 보기
myMovie = 영화평점[영화평점['mean'] >= 4.5].sort_values(by = 'count', ascending = False)
myMovie
```

```python
#3. 평점이 4.3이상이고, 평점의 수가 2000개 이상인 영화를 출력
temp = 영화평점[(영화평점['mean']>=4.3) & (영화평점['count']>= 2000)]
temp.head(15)
```

```python
temp.columns = ['평점개수','영화평점']
temp.head(15)
```
