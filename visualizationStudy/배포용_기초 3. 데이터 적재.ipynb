## 3. 데이터베이스에서 읽기
### 파이썬에서 sqlite를 내장하고 있음

```python
import sqlite3
```

```python
#테이블 생성 SQL 정의
테이블생성 = """CREATE TABLE locations
                (시 varchar[20], 구 varchar[20], 선호도 integer)"""
```

```python
#sqlite 접속
conn = sqlite3.connect('test.db')
```

```python
#테이블 생성 SQL 수행 
conn.execute(테이블생성)
```

```python
# commit() 수행. - SQL 수행 후, commit을 해야만, 데이터베이스에 SQL 구문이 적용됨 
conn.commit()
```

```python
#생성된 테이블에 저장할 데이터 정의
data = [['서울','강남', 15]
       ,['서울','강동', 10]
       ,['서울','동대문', 7]
       ,['서울','양천', 12]
       ,['성남','분당', 12]
       ,['수원','영통', 15]
       ,['성남','중원', 11]
       ,['수원','팔달', 14]
       ]
```

```python
# 데이터 적재 SQL 정의 
적재SQL = 'insert into locations values (?,?,?)'
```

```python
#데이터 적재 SQL 수행
conn.executemany(적재SQL,data)
```

```python
# commit() 수행
conn.commit()
```

```python
#Select SQL 실행1
pd.read_sql('select * from locations', conn)
```

```python
#Select SQL 실행2
pd.read_sql('select 시, sum(선호도) from locations group by 시', conn)
```

```python
# 데이터베이스의 테이블 크기가 큰 경우에는 cursor 사용하기
cursor = conn.execute('select * from locations')
```
