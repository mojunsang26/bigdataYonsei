```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

# NC 다이노스 선수 데이터 분석
## 1. 데이터 적재 및 통합

```python
rawData = pd.read_excel('data/NC Dinos.xlsx', sheetname = None )
rawData
```

## 2. 2015년도 선수 기록을 '타율'로 내림차순 정렬

```python
NC13 = rawData['2013'].set_index('선수명')
NC14 = rawData['2014'].set_index('선수명')
NC15 = rawData['2015'].set_index('선수명')
NC15.sort_values(by = '타율', ascending = False)
```

## 3.  (실습) 2013~2015년  선수별로 3년 간 평균 안타수를 구한 후, 평균 안타수가 가장 많은 상위 10명의 선수를 출력

```python
df = pd.concat([NC13,NC14,NC15], axis=0)
df2 = df.pivot_table(index = '선수명', aggfunc = 'mean', values = '안타').sort_values(by='안타',ascending = False)
players = list(df2.index[:10])
NC13
```

## 4. (실습) 3에서 구한 상위 10명의 타자들의 3년간 안타, 홈런, 타율을 출력하기

```python
안타 = pd.concat([NC13.안타,NC14.안타,NC15.안타], axis = 1)
안타.columns = [ ['안타','안타','안타'],[2013,2014,2015]]

홈런 = pd.concat([NC13.홈런,NC14.홈런,NC15.홈런], axis = 1)
홈런.columns = [ ['홈런','홈런','홈런'],[2013,2014,2015]]

타율 = pd.concat([NC13.타율,NC14.타율,NC15.타율], axis = 1)
타율.columns = [ ['타율','타율','타율'],[2013,2014,2015]]

hello = pd.concat([안타,홈런,타율], axis = 1)

hello.loc[players].fillna('-')
```

## 5. (실습) 년도별 팀의 기록 (총 안타수, 총 홈런수, 총 도루수)을 출력

```python
#NC.pivot_table(index = '년도', values = ['안타','홈런','도루'], aggfunc = 'sum')
```

```python
stat13 = NC13[['도루','안타','홈런']].sum()
stat14 = NC14[['도루','안타','홈런']].sum()
stat15 = NC15[['도루','안타','홈런']].sum()
stat = pd.concat([stat13,stat14,stat15],axis = 1).T
년도 = ['2013','2014','2015']
stat.index = 년도
stat
```
