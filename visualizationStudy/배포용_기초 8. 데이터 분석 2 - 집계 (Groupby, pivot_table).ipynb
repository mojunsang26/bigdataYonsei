```python
#실습, 자치구 별로 남자인구수의 5년치 평균을 구한 후, 평균 남자인구수가 가장 많은 상위 5개 구를 찾아보기
data.groupby('자치구')['남자인구'].mean().sort_values(ascending = False)[:5].index
```

```python
data.pivot_table(index = ['도시','자치구'], aggfunc = 'mean', values = '남자인구').sort_values(by='남자인구', ascending = False).head()
```

```python
#실습. 남자와 여자의 인구수가 가장 많이 차이나는 도시 및 자치구 찾기 (평균의 차이)
temp = data.pivot_table(index = ['도시','자치구'], aggfunc = 'mean', values = ['남자인구','여자인구'])
temp['남녀인구차이'] = abs(temp['남자인구'] - temp['여자인구'])
temp.sort_values(by = '남녀인구차이', ascending = False).head()
```

```python
# 다른 방법 (차이의 평균)
temp2 = data
temp2['차이'] = abs(data.남자인구 - data.여자인구)
temp2.pivot_table(index = ['도시','자치구'], aggfunc = 'mean', values = '차이').sort_values(by = '차이', ascending = False).head()
```
